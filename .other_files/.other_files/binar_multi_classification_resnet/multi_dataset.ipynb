```python
import shutil
import os

# Paths to the notumor folders
TRAIN_NOTUMOR = "/content/MindScan/normalized_data/Training/notumor"
TEST_NOTUMOR = "/content/MindScan/normalized_data/Testing/notumor"

# Remove the notumor folders
if os.path.exists(TRAIN_NOTUMOR):
    shutil.rmtree(TRAIN_NOTUMOR)
    print(f"Deleted: {TRAIN_NOTUMOR}")
else:
    print(f"Not found: {TRAIN_NOTUMOR}")

if os.path.exists(TEST_NOTUMOR):
    shutil.rmtree(TEST_NOTUMOR)
    print(f"Deleted: {TEST_NOTUMOR}")
else:
    print(f"Not found: {TEST_NOTUMOR}")

print("\nDone! 'notumor' folders have been removed.")
print("Your original normalized_data.zip file remains untouched in Google Drive.")
```

```python
# Rename the normalized_data folder
OLD_NAME = "/content/MindScan/normalized_data"
NEW_NAME = "/content/MindScan/multiclassification_dataset"

if os.path.exists(OLD_NAME):
    os.rename(OLD_NAME, NEW_NAME)
    print(f"Renamed folder to: {NEW_NAME}")
```

```python
# Zip the Multiclassification_dataset folder
SOURCE_FOLDER = "/content/MindScan/multiclassification_dataset"
OUTPUT_ZIP = "/content/drive/MyDrive/MindScan/multiclassification_dataset.zip"

shutil.make_archive(
    OUTPUT_ZIP.replace('.zip', ''),  # Remove .zip as make_archive adds it
    'zip',
    SOURCE_FOLDER
)

print(f"Zipped successfully to: {OUTPUT_ZIP}")
```
