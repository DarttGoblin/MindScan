```python
import os
import shutil

train_dir = "/content/drive/MyDrive/MindScan/preprocessed_data/Training"
test_dir = "/content/drive/MyDrive/MindScan/preprocessed_data/Testing"
output_dir = "/content/drive/MyDrive/MindScan/new_dataset"

classes = ["glioma", "meningioma", "notumor", "pituitary"]

# Create output class folders
for cls in classes:
    os.makedirs(os.path.join(output_dir, cls), exist_ok=True)

def copy_images(src_root, dst_root):
    for cls in classes:
        src = os.path.join(src_root, cls)
        dst = os.path.join(dst_root, cls)

        if not os.path.exists(src):
            continue

        for filename in os.listdir(src):
            src_file = os.path.join(src, filename)
            dst_file = os.path.join(dst, filename)

            if os.path.isfile(src_file):
                shutil.copy2(src_file, dst_file)

# Copy Training + Testing
copy_images(train_dir, output_dir)
copy_images(test_dir, output_dir)

print("Done. Combined dataset stored in:", output_dir)
```
