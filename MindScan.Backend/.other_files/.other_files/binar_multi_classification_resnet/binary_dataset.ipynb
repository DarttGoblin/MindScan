# Create binary dataset

```python
import shutil
import os

src_root = "../normalized_data"
binary_root = "../binary_dataset"
os.makedirs(binary_root, exist_ok=True)

for subset in ["Training", "Testing"]:
    os.makedirs(os.path.join(binary_root, subset, "tumor"), exist_ok=True)
    os.makedirs(os.path.join(binary_root, subset, "notumor"), exist_ok=True)

    tumor_classes = ["glioma", "meningioma", "pituitary"]

    for cls in tumor_classes:
        cls_dir = os.path.join(src_root, subset, cls)
        for f in os.listdir(cls_dir):
            shutil.copy(os.path.join(cls_dir, f), os.path.join(binary_root, subset, "tumor", f))

    # Copy notumor as is
    notumor_dir = os.path.join(src_root, subset, "notumor")
    for f in os.listdir(notumor_dir):
        shutil.copy(os.path.join(notumor_dir, f), os.path.join(binary_root, subset, "notumor", f))
```

```python
import os

root_train = "../binary_dataset/Training"
print('Training Folder:')
for cls in ["tumor", "notumor"]:
    count = len([f for f in os.listdir(os.path.join(root_train, cls)) if f.endswith('.npy')])
    print(f"{cls}: {count} images")

root_test = "../binary_dataset/Testing"
print('\nTesting Folder:')
for cls in ["tumor", "notumor"]:
    count = len([f for f in os.listdir(os.path.join(root_test, cls)) if f.endswith('.npy')])
    print(f"{cls}: {count} images")
```
